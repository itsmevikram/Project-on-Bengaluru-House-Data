Checking for float values

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

After splitting the ranges we trake an average of it

```python
def avg_sqft(x):
    val=x.split('-')
    if len(val)==2:
        return float(val[0])+float(val[1])/2
    try:
         return float(x)
    except:
        return None
```

```python
data3['total_sqft']=data3['total_sqft'].apply(avg_sqft)
```

Grouping out the locations and finding the sum of observation for each locations.Moreover, if sum is less than 10 obeservations each of those locations are grouped and named as 'others' 

```python
data4['location']=data4['location'].apply(lambda x: x.strip())
loc_stats=data4.groupby('location')['location'].agg('count').sort_values(ascending=False)
```

filtering away the outliers and keeping those prices which are greater tha mean -1 standard Dev. and prices which are less than equal to mean- 1 standard Dev. 

```python
import numpy as np
data_out=pd.DataFrame()
for key,subdata in data5.groupby('location'):
    pps_mean=np.mean(subdata['price_per_sqft'])
    pps_std=np.std(data5['price_per_sqft'])
    red_data=subdata[(subdata['price_per_sqft']>(pps_mean-pps_std)) & (subdata['price_per_sqft']<= (pps_mean+pps_std)) ]
    data_out=pd.concat([data_out,red_data],ignore_index=True)
data6=data_out
```

After exploring the dataset it was found that 2bhk home price is more than 3bhk home,that is quite unusual.

```python
import matplotlib
import matplotlib.pyplot as plt
def scatter_chart(df,location):
    bhk2=df[(df.location==location) & (df.BHK==2)]
    bhk3=df[(df.location==location) & (df.BHK==3)]
    matplotlib.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2['total_sqft'],bhk2['price'],marker='o',color='green',label='2BHK',s=50)
    plt.scatter(bhk3['total_sqft'],bhk3['price'],marker='x',color='black',label='3BHK',s=50)
    plt.xlabel('total sq. feet')
    plt.ylabel('price')
    plt.title('location summary')
    plt.grid(True)
    plt.legend()
```

```python
scatter_chart(data6,'Rajaji Nagar')
```

removal of those ovservations there is unusuality of bhk with prices

```python
import numpy as np
removals=np.array([])
for loc,loc_data in data6.groupby('location'):
    pps_bhk={}
    for dim,dim_data in loc_data.groupby('BHK'):
        pps_bhk[dim]={'mean':np.mean(dim_data['price_per_sqft']),
                      'std':np.std(dim_data['price_per_sqft']),
                      'count':dim_data.shape[0]
                     }
    for dim,dim_data in loc_data.groupby('BHK'):
        prev_bhk=pps_bhk.get(dim-1)
        if prev_bhk and prev_bhk['count']>5:
            removals=np.append(removals,dim_data[dim_data['price_per_sqft']<(prev_bhk['mean'])].index.values)
```

```python
data7=data6.drop(removals,axis='index')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import GridSearchCV
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.2,random_state=10)
```

finding the best Machine Learning Algorithm along with its Parameterswhich is suitable for the dataset using cross validation technique

```python
def find_best_score(x,y):
    model_grid={'LinearRegression':{
        'model':LinearRegression(),
        'parameters':{
            'normalize':[True,False]
        }
    },
            'Lasso':{
                'model':Lasso(),
                'parameters':{
                    'alpha':[1,2],
                    'selection':['random','cyclic']
                }
            },
            'DecisionTreeRegressor':{
                'model':DecisionTreeRegressor(),
                'parameters':{
                    'criterion':['mse','friedman_mse'],
                    'splitter':['best','random']
                }
            }
           }
    model_scores=[]
    cross_validation=ShuffleSplit(n_splits=5,test_size=0.2)
    for model_name,mdl in model_grid.items():
        grid_search=GridSearchCV(mdl['model'],mdl['parameters'],cv=cross_validation,return_train_score=False)
        grid_search.fit(X , y)
        model_scores.append({
            'model':model_name,
            'score':grid_search.best_score_,
            'parameters':grid_search.best_params_
        })
    return pd.DataFrame(model_scores,columns=['model','score','parameters'])
```

```python
find_best_score(X,y)
```

```python
lr_model=LinearRegression()
lr_model.fit(xtrain,ytrain)
```

Prediction parameters arrangements

```python
def model_predict(loc,Sq_ft,bath,BHK):
    index=np.where(X.columns==loc)[0][0]
    a=np.zeros(len(X.columns))
    a[0]=Sq_ft
    a[1]=bath
    a[2]=BHK
    if index >=0:
        a[index]==1
    return lr_model.predict([a])[0]
```
